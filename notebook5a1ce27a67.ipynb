```python
import sys
import os
import time

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras import activations, backend as K

from keras.preprocessing.sequence import pad_sequences
from keras.applications.resnet import *
from keras.applications.efficientnet import *
from keras.applications.mobilenet import *
from keras.applications.mobilenet_v2 import *
from keras.applications.vgg16 import *
from keras.applications.vgg19 import *
from keras.applications.densenet import *
from keras.models import load_model
from keras.metrics import Precision, Recall

from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns

from IPython import display
from PIL import Image
import random
import json
import datetime
```

```python
base_model = VGG19(
    weights='imagenet',
    include_top=False, 
    classes = 64
)

model = keras.Sequential()

# Input
model.add(layers.Input(shape=(224, 224, 3), dtype=tf.uint8))
# VGG
model.add(base_model)

model.add(layers.Flatten())
# Output
model.add(layers.Embedding(10000, 64, name="embedding_norm")) # add an Embedding layer

model.compile()
```

```python
model.save('saved_model')
model.summary()
```
