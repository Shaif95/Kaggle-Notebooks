```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob
```

```python
# PATHS
TRAIN_IMAGES = "../input/mayo-clinic-strip-ai/train/*.tif"
TEST_IMAGES = "../input/mayo-clinic-strip-ai/test/*.tif"
TRAIN_CSV = "../input/mayo-clinic-strip-ai/train.csv"
TEST_CSV = "../input/mayo-clinic-strip-ai/test.csv"
SAMPLE_SUB_CSV = "../input/mayo-clinic-strip-ai/sample_submission.csv"
OTHERS_CSV = "../input/mayo-clinic-strip-ai/other.csv"
```

```python
m = []
a = glob.glob(TEST_IMAGES)
for i in a :
    x = i.split('/')
    l = x[4].split('.')[0].split('_')[0]
    m.append(l)
    
#m
df = pd.DataFrame(m, columns=['patient_id'])
r = len(glob.glob(TEST_IMAGES))
arr = []
arr = [.5 for i in range(r)] 
df['CE'] = arr
df['LAA'] = arr
df = df.groupby('patient_id').mean()
df.head(10)
```

```python
df.to_csv('submission.csv')  
```
